# College Football bowl games Analysis
In this project, a CSV file with medical insurance costs will be investigated using Python.
The goal with this project will be to analyze various attributes within collegefootballbowls.csv to learn
more about the bowl information in the file.

```python
#import csv file
import csv
#import Statistcs Module
import statistics
#import matplotlib Module
import matplotlib.pylab as plt
```

```python
#Create empty lists for some of the attributes in the csv file
years = []
day = []
winner_points = []
loser_points = []
```

```python
#Helper function to load csv file
def load_list_data(lst, csv_file, column_name):
    # open csv file
    with open('collegefootballbowl.csv') as csv_info:
        # read the data from the csv file
        csv_dict = csv.DictReader(csv_info)
        # loop through the data in each row of the csv 
        for row in csv_dict:
            # add the data from each row to an empty list
            lst.append(row[column_name])
        # return the list
        return lst
```

```python
# look at the data in insurance_csv_dict
load_list_data(years, 'collegefootballbowls.csv', 'year')
load_list_data(day, 'collegefootballbowls.csv', 'day')
load_list_data(winner_points, 'collegefootballbowls.csv', 'winner_points')
load_list_data(loser_points, 'collegefootballbowls.csv', 'loser_points')
```

Now the data from collegefootballbowls.csv is neatly organized into labeled lists
The following operations will be implemented:

* Find the number of bowl games per year
* Find the day of the week in which most bowl games are played
* Find the mean, median, and standard deviation for the score of winning teams
* Find the mean, median, and standard deviation forr the score of losing teams.

To perform these inspections, a class called Bolws has been built out which contains fives methods:

* bowls_per_year()
* day_of_game()
* w_points()
* l_points()

The class has been built out below.

```python
class Bowls:
    # init method that takes in each list parameter
    def __init__(self, year, day, winner_points, loser_points):
        self.year = year
        self.day = day
        self.winner_points = winner_points
        self.loser_points = loser_points
    

    # method that calculates the number of games per year
    def bowls_per_year(self):
        #create empty dictionary
        year_dic = {}
        #loop thad adds each year there was a bowl game as a key to the dicitonary
        #the values of each key are the number of games played in that year
        for element in self.year:
            if element in year_dic:
                year_dic[element] += 1
            else:
                year_dic[element] = 1

        # myList access the key-value pairs in the dictionary
        myList = year_dic.items()
        #sorted function so that graph is in order
        myList = sorted(myList)
        #values are extracted from the list using the zip function
        x, y = zip(*myList)
        #plt.plot function plots the values
        plt.plot(x, y)
        #x axis
        plt.xlabel('Year')
        #y axis
        plt.ylabel('Number of games')
        #title of graph
        plt.title('Bowl Games played Each Year')
        #shows graph
        plt.show()
    
    #method that calculates how many bowl games were played each day of the week
    def day_of_game(self):
        #create empty dictionary
        day_played = {}
        #loop that adds each day of the week as a key
        #the values of each key is how many games were played in that day of the week
        for day in self.day:
            if day in day_played:
                day_played[day] += 1
            else:
                day_played[day] = 1
         
        #sums the total number of games played
        day_values = day_played.values()
        total_days = sum(day_values)
        
        #loop through a dictionary and prints the number of games played each day of the week
        for key, value in day_played.items():
            print("The percentage of games played on a {key} is {value} %.".format(key = key, value = round(value/total_days * 100,2)))
            
            
        
    #method that calculates analyzes the score of winning teams
    def w_points(self):
        #list comprehension to convert a list of strings to a list of integers
        winner_points_lst = [int(point) for point in self.winner_points]
        #mean, median and stdev were calculated using the statistics modules
        mean_points = statistics.mean(winner_points_lst)
        median_points = statistics.median(winner_points_lst)
        stdev_points = statistics.stdev(winner_points_lst)
        print("The mean for the score of teams that won is {}".format(mean_points))
        print("The median for the score of teams that won is {}".format(median_points))
        print("The standard deviation for the score of teams that won is {}".format(stdev_points))
        
    def l_points(self):
        #list comprehension to convert a list of string to a list of integerrs
        loser_points_lst = [int(point) for point in self.loser_points]
        #mean, median, and stdev were calcualted using the statistcs module
        mean_points = statistics.mean(loser_points_lst)
        median_points = statistics.median(loser_points_lst)
        stdev_points = statistics.stdev(loser_points_lst)
        print("The mean for the score of teams that lost is {}".format(mean_points))
        print("The median for the score of teams that lost is {}".format(median_points))
        print("The standard deviation for the score of teams that lost is {}".format(stdev_points))
        
        
    
    
```

The instance variable called bowl_info was created. With this variable the methods can be used

```python
bowl_info = Bowls(years,day, winner_points, loser_points)
```

```python
bowl_info.bowls_per_year()
```
